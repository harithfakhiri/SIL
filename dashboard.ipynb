```python
#importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Encoding categorical data values

#importing our cancer dataset
dataset = pd.read_csv('./keuangan.csv')

dataset = dataset.drop(['kategori_penilaian'], axis=1)
```

```python
index_year = dataset.groupby(['tahun']).sum().sort_values(by="nilai_index", ascending=True)
index_year.plot(kind='line', y='nilai_index', figsize=(15,6), xticks=dataset['tahun'])
```

```python
index_year = dataset.groupby(['nama_kabupaten_kota']).sum().sort_values(by="nilai_index", ascending=False)
index_year.plot(kind='barh', y='nilai_index', figsize=(20,8))
```

```python
sort_kabupaten = dataset.sort_values(["tahun", "nilai_index"], ascending=False)
sort_kabupaten[sort_kabupaten['tahun']==2020].iloc[:5].plot(kind='bar', x="nama_kabupaten_kota", y='nilai_index', figsize=(9,6), title="tahun 2020")

sort_kabupaten[sort_kabupaten['tahun']==2019].iloc[:5].plot(kind='bar', x="nama_kabupaten_kota", y='nilai_index', figsize=(9,6), title="tahun 2019")

sort_kabupaten[sort_kabupaten['tahun']==2018].iloc[:5].plot(kind='bar', x="nama_kabupaten_kota", y='nilai_index', figsize=(9,6), title="tahun 2018")
```

```python
new_set = df = pd.DataFrame()
new_set['kode_kabupaten_kota'] = dataset['kode_kabupaten_kota'].iloc[:27]
new_set['nama_kabupaten_kota'] = dataset['nama_kabupaten_kota'].iloc[:27]
a = []
for b in new_set['kode_kabupaten_kota']:
    temporary = dataset[dataset['kode_kabupaten_kota']==b]
    temp = temporary[temporary['tahun']==2020]['nilai_index'].max() - temporary[temporary['tahun']==2018]['nilai_index'].min()
    # print(temp)
    a.append(temp)
a = np.array(a)
new_set['growth'] = a.tolist()
```

```python
new_set.plot(kind='barh', y="growth", x='nama_kabupaten_kota', figsize=(15,10), title="perkembangan nilai index tahun 2018 - 2020")
```
